## Cargamos los datos a nuestra BBDD de SQL

```python
def check(que,string):
    """
    Función parametrizada que comprueba en cada tabla si existe el user, artista o canción.
    Devuelve True si existe y False si no
    """
    if que == "Season":
        query = list(engine.execute(f"SELECT Number FROM Season WHERE Number = '{string}'"))
        if len(query) > 0:
            return True
        else:
            return False
        
    elif que == "Episode":
        query = list(engine.execute(f"SELECT Name FROM Episode WHERE Name = '{string}'"))
        if len(query) > 0:
            return True
        else:
            return False
       
    elif que == "Actors":
        query = list(engine.execute(f"SELECT Name FROM Actors WHERE Name = '{string}'"))
        if len(query) > 0:
            return True
        else:
            return False
        
    elif que == "Dialogue":
        query = list(engine.execute(f"SELECT Dialogue FROM Dialogue WHERE Dialogue = '{string}'"))
        if len(query) > 0:
            return True
        else:
            return False
```

```python
check("Episode", "Hola")
```

```python
def insertSeason(string):
    """
    Llama a la función check para comprobar si existe season
    Inserta season si no existe
    """
    if check("Season", string):
        return "Esa temporada ya existe"
    else:
        engine.execute(f"INSERT INTO Season (Number) VALUES ('{string}');")
```

```python
def insertEpisode(nombre, number):
    """
    Llama a la función check para comprobar si existe el Episode
    Inserta episode si no existe
    """
    if check("Episode", nombre):
        return "El Episodio ya existe"
    else:
        engine.execute(f"""INSERT INTO Episode (Name, Number_episode, Season_id_Season) 
                       VALUES ('{nombre}', '{number}', 1);""")
```

```python
def insertActors(nombre):
    """
    Llama a la función check para comprobar si existe el Actor
    Inserta actor si no existe
    """
    if check("Actors", nombre):
        return "El Actor ya existe"
    else:
        engine.execute(f"INSERT INTO Actors (Name) VALUES ('{nombre}');")
```

```python
def dameId(que,string):
    """
    Devuelve el ID de lo que le pidamos sabiendo que ese elemento EXISTE
    """
    if que == "Season":
        return list(engine.execute(f"SELECT id_Season FROM Season WHERE Number ='{string}';"))[0][0]
    elif que == "Episode":
        return list(engine.execute(f"SELECT id_episode FROM Episode WHERE Name ='{string}';"))[0][0]
    elif que == "Actors":
        return list(engine.execute(f"SELECT id_Actors FROM Actors WHERE Name ='{string}';"))[0][0]
```

```python
def insertDialogue(fila):
    """
    La función final 
    """
    if check("Dialogue", fila["Dialogue"]):
        return "El dialogo ya existe"
    else:
        if check("Episode", fila["Name_Episode"]):
            episode_id = dameId("Episode", fila["Name_Episode"])
        else:
            insertEpisode(fila["Name_Episode"], fila["Number_Episode"])
            episode_id = dameId("Episode", fila["Name_Episode"])
        
        if check("Actors", fila["Actor"]):
            actors_id = dameId("Actors", fila["Actor"])
        else:
            insertActors(fila["Actor"])
            actors_id = dameId("Actors", fila["Actor"])
    
        engine.execute(f"""
        INSERT INTO Dialogue (Episode_id_episode, Actors_id_Actors, Dialogue) 
        VALUES ("{episode_id}", "{actors_id}", "{fila['Dialogue']}");
        """)
```
